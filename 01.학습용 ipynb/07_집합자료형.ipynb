### 집합(set)
중복을 허용하지 않는다
순서가 없다.


활용 예 : 중복 검사, 리스트 업 할 때(ex. 출석부)

그럼 리스트에 모든 요소 집어 넣고 find()로 중복 검사하면 되는 것 아닌가?

: 리스트는 순서가 있기 때문에 요소가 많아질수록 검사 시간이 오래 걸린다.
집합은 순서가 없기 때문에 요소가 많아져도 검사 시간이 오래 걸리지 않는다.
find()는 주소를 순서대로 찾기 때문...

```python
s0 = {"a","b","c","a"}
s1 = set([5,1,3,3,6])
s2 = set("hello")
print(s0)
print(s1)
print(s2)
```

```python
s = set("hello")
print("l"in s)
```

```python
# 교집합, 합집합, 차집합
s1 = set([1,2,3,4,5,6])
s2 = set([4,5,6,7,8,9])
print("교집합")
print(s1 & s2)
print(s1.intersection(s2))
print("합집합")
print(s1 | s2)
print(s1.union(s2))
print("차집합")
print(s1-s2)
print(s1.difference(s2))
```

```python
# 집합은 값 업데이트 가능
s = set([1,2,3])
s.add(4)
print(s)

# 여러 값 추가하기
s.update([4,5,6 ])
print(s)
s.update(set([7,8,9]))
print(s)

# 값 지우기
s.remove(9)
print(s)
```
