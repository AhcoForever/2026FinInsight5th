```python
# 1. 부모 태그와 선택자로 넓은 범위에서 접근
company = soup.select('div.col.company_nm')
len(company)
```

```python
company
```

```python
# 2. 원하는 태그에 각각 접근하기
# (여러 태그가 복잡하게 얽혀 있을 조건문으로 경우 상황에 맞게 태그 값을 지정해서 정보를 가져오기)
company_list = []

for i in company :
    # a태그에 값이 있다면 True로 인지함
    if i.select('a') :
        # select로 해도 되지만 리스트로 반환 받으므로 한 번 더 정제해야하는데 select_one은 하나의 값만 반환해서 편하게 사용가능
        # (어차피 우리가 원하는 회사명은 단일 값이므로 select_one으로 받는게 더 효율적)
        target = i.select_one('a')
        
    # a태그가 없다면 span 태그를 확인
    elif i.select('span') :
        target = i.select_one('span')

    company_list.append(target.text.strip())
company_list
```

## 이제 1~10page의 데이터를 모두 수집해보자!
- page의 경우 GET방식의 웹사이트는 url에 정보가 표시되므로 해당 정보가 어디있는지 파악해 볼 것.

```python
# tqdm : 반복문의 진행상황을 로딩바로 확인하게 해주는 모듈
from tqdm import tqdm
```

```python
titles_list, authors_list, views_list = [], [], []

for i in tqdm(range(1,11,1)) :
    # 1~10page에 반복적으로 접근
    url = f'https://www.10000recipe.com/recipe/list.html?order=reco&page={i}'
    res = req.get(url, headers=U_A)
    soup = bs(res.text, 'lxml')

    titles = soup.select('div.common_sp_caption_tit.line2')
    authors = soup.select('div.common_sp_caption_rv_name > a')
    views = soup.select('span.common_sp_caption_buyer')

    for i in range(len(titles)) :
        titles_list.append(titles[i].text.strip())
        authors_list.append(authors[i].text.strip())
        views_list.append(views[i].text.strip()[4:])

recipe_dict = {'레시피명':titles_list, '작성자':authors_list, '조회수':views_list}
recipe_df = pd.DataFrame(recipe_dict, index=range(1,len(titles_list)+1))
recipe_df.index.name = 'No'
recipe_df
```

### 특정 키워드가 들어간 레시피만 필터링

```python
filtered_df = recipe_df[recipe_df['레시피명'].str.contains('김치')]
# reset_index : 인덱스 재설정
 # drop=True : 기존 인덱스 컬럼 삭제
filtered_df.reset_index(drop=True,inplace=True)
filtered_df.index = range(1, len(filtered_df)+1)   # 필터링 된 데이터프레임 행 개수만큼 인덱스 설정
filtered_df.index.name = 'No'
filtered_df
```

### 함수로 코드 통합하기

```python
# 매개변수는 검색할 최종페이지, 필터링 할 검색어(디폴트로 빈 문자열)
# 반환할 자료형은 DataFrame을 명시
def search_recipe(end_page:int, search:str='') -> pd.DataFrame :
    titles_list, authors_list, views_list = [], [], []

    for i in tqdm(range(1, end_page+1, 1)) :
        url = f'https://www.10000recipe.com/recipe/list.html?order=reco&page={i}'
        res = req.get(url, headers=U_A)
        soup = bs(res.text, 'lxml')
    
        titles = soup.select('div.common_sp_caption_tit.line2')
        authors = soup.select('div.common_sp_caption_rv_name > a')
        views = soup.select('span.common_sp_caption_buyer')
    
        for i in range(len(titles)) :
            titles_list.append(titles[i].text.strip())
            authors_list.append(authors[i].text.strip())
            views_list.append(views[i].text.strip()[4:])
    
    recipe_dict = {'레시피명':titles_list, '작성자':authors_list, '조회수':views_list}
    recipe_df = pd.DataFrame(recipe_dict, index=range(1,len(titles_list)+1))

    # 검색어가 존재할 경우
    if search :
        # copy : 기존 객체를 복사
        filtered_df = recipe_df[recipe_df['레시피명'].str.contains(f'{search}')].copy()
    # 검색어가 없으면
    else :
        # 기존 recipe_df를 그대로 사용
        filtered_df = recipe_df.copy()

    filtered_df.reset_index(drop=True,inplace=True)
    filtered_df.index = range(1, len(filtered_df)+1)
    filtered_df.index.name = 'No'
    return filtered_df
```

```python
search_recipe(5, '김치')
```
