### while 문

```python
# a를 1씩 증가하며 출력하기: 5보다 작을 동안
a= 0
print(f"out: {a}")
while a<5:
    print(f"in:{a}")
    a+=1
print(f"out: {a}")
```

```python
# 입력 값 받기
prompt = """
1. Add / 2. Del / 3. List / 4. Quit
"""

number = 0
while number !=4:
    print(prompt)
    number = int(input()) # 사용자에게 값 받기
```

```python
# 피보나치 수열
fibo = [1,1]
index = 0
while index <10:
    fibo.append(fibo[index]+ fibo[index+1])
    index +=1
print(fibo)
```

```python
## 99까지 3, 6, 9 출력하기
number = 1 # 상수는 일반적으로 대문자로 명명한다
MAX = 99
while number <= MAX:
    if number % 10 in [3, 6, 9] or number // 10 in [3, 6, 9]:
        print("**", end=", ")
    else:
        print(f"{number:<2}", end=", ")
    if number % 10 == 0:
        print("") # 10번 찍을 때마다 줄바꿈
    number +=1

    # 다른 방법 : 숫자를 문자로 바꾸고 find로 바꾸면 찾기 쉬워진다.

```

```python
# 중간에 흐름 제어하기
# continue : 뒷부분 수행 하지 않고 다시 위로 
num = 0
while num <10:
    num +=1
    if num %3 ==0:
        continue
    print(f"in:{num}")
```
