### 함수

```python
# 입력 값에 따라 일관 적인 결과 값을 내놓는 코드 정의

# def 함수_이름(매개변수):
#   수행문
#   수행문
#   return 반환 값 
```

```python
# f(x) = 2x + 1
def f(x):
    y = 2 * x + 1
    return y

i = 10
result = f(i)
print(f"f({i}) = 2 * {i} + 1 = {result}")
i = 20
result = f(i)
print(f"f({i}) = 2 * {i} + 1 = {result}")
```

```python
# 매개변수는 필수가 아님
# 반환 값도 필수가 아님
def say_hello():
    print("hello")

say_hello()
```

```python
# 매개 변수는 정의한 순서대로 입력 필요

def add(a, b):
    print(f"{a} + {b} = {a + b}")

def sub(a, b):
    print(f"{a} - {b} = {a - b}")

add(5, 3)
sub(5, 3)
```

```python
# 매개 변수는 정의한 순서대로 입력 필요
# 호출 시, 명시적으로 매개변수에 값 입력 가능
def sub(a, b):
    print(f"{a} - {b} = {a - b}")

sub(3, 5)
sub(b = 3, a = 5)
```

```python
# 리스트에서 최대 값 찾기

def get_high(inputs):
    result = float('-inf')
    for i in inputs:
        if result < i:
            result = i
    return result

inputs = [3, 40, 20, 15, 60]
high = get_high(inputs)
print(f"high: {high}")
```

```python
# 매개변수의 수 여러개
# 앞에 * 붙이기

def add(*args):
    result = 0
    for i in args:
        result = result + i
    return result

result = add(1, 2, 3)
print(result)
result = add(1, 2, 3, 4, 5, 6)
print(result)
```

```python
# 키워드 매개변수
# 매개변수 내용 자체를 dict로 받기

def f(**kwargs):
    print(kwargs)

f(a = 3, b = 10)
```

```python
# 매개변수의 초기값 주기
# 입력하지 않으면 초기값으로 대응

def get_some(inputs, high = True):
    if high:
        result = float('-inf')
        for i in inputs:
            result = i if i > result else result
    else:
        result = float('inf')
        for i in inputs:
            result = i if i < result else result
    return result

inputs = [1, 2, 3, 4, 5]
r1 = get_some(inputs)
r2 = get_some(inputs, high = True)
r3 = get_some(inputs, False)
r4 = get_some(inputs, high = False)

print(r1, r2, r3, r4)
```

```python
# return, 반한 되는 시점에 함수는 중단

def f(a, b):
    return a + b
    return a * b

print(f(3, 5))
```

```python
def f():
    for i in range(100):
        if i == 10:
            return
        print(i)

f()
```

### 변수의 범위

```python
a = 1
def f(a):
    print(f'=== before: {a}')
    a = a + 1
    print(f'=== after: {a}')

print(f'before: {a}')
f(a)
print(f'after: {a}')
```

```python
# 상단 Restart 후 실행해보세요

def f(a):
    print(f'=== before: {a}')
    a = a + 1
    print(f'=== after: {a}')

f(3)
print(f'after: {a}')
```

```python
# return 값 활용

a = 1
def f(a):
    print(f'=== before: {a}')
    a = a + 1
    print(f'=== after: {a}')
    return a

print(f'before: {a}')
a = f(a)
print(f'after: {a}')
```

```python
# global 값 활용

a = 1
def f():
    global a
    print(f'=== before: {a}')
    a = a + 1
    print(f'=== after: {a}')

print(f'before: {a}')
f()
print(f'after: {a}')
```

```python
# lambda: 간결한 함수 정의 시 활용
# lambda [매개변수]: [반환 값]
f = lambda a, b: a + b
result = f(3, 4)
print(result)
```
