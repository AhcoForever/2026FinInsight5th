### 파일 입출력

```python
f = open("new.txt","w", encoding="utf-8")
f.close()
# open과 close는 반드시 한 쌍처럼 함께 사용해야함!
```

```python
# 없는 파일을 "읽기 모드"로 열면 오류 발생
f = open("hello.txt", "r", encoding="utf-8")
f.close()
```

```python
# 시스템 관련 라이브러리
import os

# 해당 파일, 경로가 존재하는지 확인
if os.path.exists("hello.txt"):
    print("파일 있어요!")
    f = open ("hello.txt", "r", encoding="utf-8")
    f.close()
else:
    print("파일 없어요!")
```

```python
# 파일 데이터 쓰기
# 쓰기모드는 새로 파일을 씁니다 -> 기존 데이터 삭제
f = open("hello.txt", "a", encoding="utf-8")
for i in range(1,6):
    data = f"{i}번째 줄입니다.\n"
    f.write(data)
    
    dataset=[]
for i in range(6,10):
     dataset.append(f"{i}번째 줄입니다. \n")
f.writelines(dataset)
f.close()
```

```python
# with open 으로 코드를 짜면 f.code()를 안써도 됨 (더 간편하게 코딩 가능)
with open("hello.txt", "w", encoding="utf-8") as f:
    for i in range(1, 5):
        data = f"{i}번째 줄입니다.\n"
        f.write(data)
```

```python
# 파일 읽기 - 첫번째 줄 읽기
f = open("hello.txt","r", encoding="utf-8")
line = f.readline()
print(line)
f.close()
```
