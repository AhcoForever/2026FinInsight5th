### 리스트

```python
# 리스트 길이
l = [1,2,3]
len(l)
```

```python
# 리스트 인덱싱
l = [1,2,3]
l[0] + l[1]
```

### 리스트 연산

```python
# 리스트 더하기
l1 = [1,2,3]
l2 = [4,5,6]
l1+l2
```

```python
# 리스트 반복하기
l = [1,2,3]
l *3
```

```python
# 리스트 수정
l[2] = 4
print(l)
```

```python
# 리스트 요소 삭제
del l[2]
print(l)
```

### 리스트 함수

```python
# append : 뒤에 값 추가하기
l = [1,2,3]
l.append(4)
print(l)
```

```python
# sort : 정렬하기  - 오름차순
l =[5,6,3,1]
l.sort()
print(l)
```

```python
# sort : 정렬하기 - 내림차순
l = [5,6,3,1]
l.sort(reverse=True)
print(l)
```

```python
# insert: 특정 위치에 값 삽입하기.
l = [1,2,3]
# insert(i, V)
# i : index 
# V : value

l.insert(1, 2)
print(l)
```

```python
# remove: 해당 값 삭제 (앞에서 부터 1개만)
l = [1,2,3,1,2,3]
l.remove(1)
print(l)
```

```python
# pop: 맨 마지막 값 꺼내기
l = [1,2,3]
print(l.pop()) # 맨 마지막 요소 꺼내서 "보여준다"
print(l)
# pop: 사전적 의미는 '튀다' 프로그래밍 언어에서는 "꺼낸다"
```

```python
# extend : 리스트 확장
l = [1,2,3]
l.extend([4,5])
print(l)
```

```python
# 리스트 복사
l1 = [1,2,3]
l2 = l1[:]

print("변경 전")
print(l1)
print(l2)

l1[0] = 4
print(l1)
print(l2)
print("=" *10)
print(f"l1 : {id(l1)}")
print(f"l2: {id(l2)}")
# id : 메모리 주소 값
```

```python
# 리스트 복사
l1 = [1,2,[3,4]]
l2 = l1[:]

print("변경 전")
print(l1)
print(l2)
l1 [2][0] = 4
print("변경 후")
print(l1)
print(l2)

print(f"l1: {id(l1[2])}")
print(f"l2: {id(l2[2])}")
```
