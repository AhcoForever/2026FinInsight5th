```python
# [미션 1] 데이터 불러오기 및 훑어보기

# korea_movie_best.csv 파일을 불러와 movie_df 변수에 저장하세요.
# 데이터가 잘 들어왔는지 상위 5개 행을 출력해서 확인하세요.
# 데이터의 행/열 개수와 컬럼별 타입(info)을 확인하세요.

movie_df = pd.read_csv('korea_movie_best.csv')
print(movie_df.head())
print("-" * 20)
print(movie_df.info())
```

```python
# [미션 2] 데이터 선택하기 (Indexing)

# 영화 제목(영화명)을 인덱스로 지정(set_index)하여 새로운 변수 df_index에 저장하세요.
# loc을 사용하여 영화 **'기생충'**의 **'평점'**을 찾아 출력하세요.

df_index = movie_df.set_index('영화명')
# '기생충' 행의 '평점' 열 선택
print("\n기생충 평점:", df_index.loc['기생충', '평점'])
```

```python
# [미션 3] 조건으로 필터링하기 (Boolean Indexing)

# 관객수가 1400만 명을 넘는 초대박 영화만 골라내서 hit_movie 변수에 저장하세요.
# 장르가 **'코미디'**인 영화만 골라내서 출력하세요.

hit_movie = movie_df[movie_df['관객수'] > 1400]
print("\n--- 1400만 이상 관객 영화 ---")
print(hit_movie)

# 장르가 코미디
print("\n--- 코미디 영화 ---")
print(movie_df[movie_df['장르'] == '코미디'])
```

```python
# [미션 4] 데이터 정렬 및 파생변수 (Sorting & Calculation)

# 전체 영화를 평점이 높은 순서대로(내림차순) 정렬해서 1위부터 3위까지만 출력하세요.
# 영화 티켓 가격이 10,000원이라고 가정합니다. 관객수 * 10,000 을 계산하여 매출액이라는 새로운 컬럼을 만드세요.
# (주의: 관객수 단위가 '만 명'이므로, 관객수 * 10000 * 10000을 해야 정확한 원화 금액이 됩니다. 수업 때는 간단히 관객수 * 10000으로 실습해도 무방합니다.)

# 평점 높은 순 정렬 후 상위 3개(head(3))
top3_rating = movie_df.sort_values(by='평점', ascending=False).head(3)
print("\n--- 평점 TOP 3 ---")
print(top3_rating)

# 매출액(단위: 억) 컬럼 추가 (관객수 * 1억 / 10000 등의 복잡한 계산 대신 단순화)
# 여기서는 간단히 관객수(만명) * 10000원으로 계산해봅니다.
movie_df['매출액'] = movie_df['관객수'] * 10000
print("\n--- 매출액 컬럼 추가 결과 ---")
print(movie_df[['영화명', '관객수', '매출액']].head())
```
