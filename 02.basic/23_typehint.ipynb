```python
# 함수 f는 어떤 함수 인가요?
# f에 마우스 가져다 대보기

def f(a,b):
    return a+b

f(3,5)
f("a", "b")
```

```python
# 변수 및 리턴 타입에 대하여 표기 가능
# f에 마우스 가져다 대보기
# 강제성은 없음
# 강제성을 주려면 extention 설치해야함.
def f(a:int, b:int) -> int:
    """
    a와 b를 더하는 함수
    """
    return a+b

name:str = "hello"
age:int = 20
```
