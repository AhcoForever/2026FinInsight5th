### 튜플
()안에 한번 선언하면 내부 요소를 변경할 수 없다. 
이터러블

```python
t1 = ()
t2 = (1,) # (1) 라고 적을 경우 숫자로 인식되기 때문에 단일 숫자를 튜플로 인식 시켜 주기 위해서 , 를 추가해야한다.
t3 = (1,2,3)
t4 = 1,2,3
t5 = (1,2,(3,4))
```

```python
# index 접근
t = (1,2,3)
print(t[0])
print(t[1:])
# 출력 형식이 다름
```

```python
# 변수 선언
(a,b) = (1,2)
# 튜플을 여러 방식으로 선언해 줄 수 있다. 
```
