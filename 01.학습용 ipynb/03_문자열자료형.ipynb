### 포매팅

```python
"저는 %d년생 입니다." % 2000
```

```python
"제 이름은 %s이고, 나이는 %d살 입니다." % ("김민영",20)
```

```python
# 정렬과 공백
"===%10s===" %"PSH"
# 오른쪽 정렬, 10칸 공백, 왼쪽 정렬은 %-10s 라고 표현한다.
```

```python
# 소수점 표현
"%.4f" % 3.14159265 
# 소수점 4자리까지만 표현
```

```python
# format 함수 사용하기
# .format 내 값 나열, 앞쪽에 0번부터 위치 나열
print("이것은 {0}번째 레슨".format(1))
print("제 이름은 {0}이고, 나이는 {1}입니다.".format("김민영",20))
```

```python
# f 문자열 포매팅 : 파이썬 3.6 이상부터 지원
name = "김민영"
age = 20
f'제 이름은 {name}이고, 나이는 {age}입니다.'
```

### 문자열 함수

```python
# count : 문자열 내 해당 단어 숫자 세기
ex = "starbucks"
print("s: %d" %(ex.count('s')))
print("star: %d" % (ex.count('star')))
```

```python
# find : 문자열 내 해당 단어 첫 위치 찾기
# 없으면 -1
ex = "starbucks"
print("s: %d" %(ex.find('s')))
print("z: %d"%(ex.find('z')))
```

```python
# index : 문자열 내 해당 단어 첫 위치 차직
# 없으면 error
ex = "starbucks"
print("s: %d" %(ex.index('s')))
print("z: %d" %(ex.index('z')))
```

```python
# lstrip : 왼쪽 공백 지우기
# restrip : 오른쪽 공백 지우기
# strip : 양쪽 공백 지우기
ex = " abc"
print("[" + ex.lstrip() +"]")
print("["+ exrstrip(+)"]")
```

```python
# join: 문자열 각 사이에 해당 문자 삽입
print(" ".join("starbucks"))
print("/".join("starbucks"))
```

```python
# split : 특정 문자 기준으로 나누기 => 나눈 결과는 List로 반환
ex = 's t a r b u k s'
# 비어있으면 공백 기준으로 나눔
ex.split()
```

```python
# split 나눌 기준이 존재하지 않으면...
ex='abcdefg'
ex.split('Z')
```
