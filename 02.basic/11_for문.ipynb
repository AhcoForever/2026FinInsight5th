### for문

```python
# 일정한 동작 반복 수행이 필요할 때 활용하는 반복문
# 주어진 범위 내에서 반복 수행
# for [변수] in [범위: 리스트, 튜플, 문자열, ...]:
#   [수행]
```

```python
# 과일 리스트 출력하기
a = ['apple', 'banana', 'orange']
for f in a:
    print(f)
```

```python
# 문자열 접근
for c in "hello":
    print(c)
```

```python
# 리스트 내 튜플 처리
t = [(1, 2), (3, 4), (5, 6)]
for (a, b) in t:
    print(a + b)
```

### range 함수

```python
# range 함수
# 숫자 리스트를 자동으로 만들어주는 함수

r = range(5)          # range(N): 0이상 N미만의 숫자 리스트
for i in r:
    print(i)
```

```python
# range 함수
# 숫자 리스트를 자동으로 만들어주는 함수

r = range(1, 5)          # range(s, N): s이상 N미만의 숫자 리스트
for i in r:
    print(i)
```

```python
# range 함수
# 숫자 리스트를 자동으로 만들어주는 함수

r = range(1, 5, 2)          # range(s, N, g): s이상 N미만의 숫자 리스트 g씩 건너뛰기
for i in r:
    print(i)
```

```python
# range 함수
# 숫자 리스트를 자동으로 만들어주는 함수

a1 = range(5, -1, -1)          # 내림차순
for i in a1:
    print(i)
```

```python
# range 활용 과일 리스트 출력하기
a = ['apple', 'banana', 'orange']
for i in range(len(a)):
    print(a[i])
```

```python
# 남성, 여성 회원 수 세기
members = [
    { 'name': 'A', 'gender': 'M'},
    { 'name': 'B', 'gender': 'M'},
    { 'name': 'C', 'gender': 'F'},
    { 'name': 'D', 'gender': 'M'},
    { 'name': 'E', 'gender': 'F'},
]

m_cnt = 0
f_cnt = 0
for m in members:
    if m['gender'] == 'M':
        m_cnt += 1
    else: 
        f_cnt += 1
print(f"M: {m_cnt} / F: {f_cnt}")
```

```python
# 피보나치 수열
N = 10
fibo = [1, 1]
for i in range(2, N):
    fibo.append(fibo[i - 2] + fibo[i - 1])
print(fibo)
```

### 다중 for문

```python
# for문 내에서 또 for문을 활용하여 프로그래밍
# i loop 반복 수 * j loop 반복 수 만큼 실행
for i in range(3):
    print(f"=== i : {i}")
    for j in range(5):
        print(f"j: {j}")
```

```python
# 다중 for문 활용 구구단 출력하기
for i in range(2, 10):
    print(f"[{i}단]")
    for j in range(1, 10):
        print(f"{i} x {j} = {i * j}")
```

### 별 찍기

```python
# 별찍기: 정사각형
N = 5
for i in range(N):
    for j in range(N):
        print("*", end="")
    print("")
```

```python
# 별찍기: 정사각형
N = 5
for i in range(N):
    print("*" * N)
```

```python
# 별찍기: 직각삼각형
N = 5
for i in range(N):
    for j in range(i + 1):
        print("*", end="")
    print("")
```

```python
# 별찍기: 직각삼각형
N = 5
for i in range(N):
    print("*" * (i + 1))
```

```python
# 별찍기: 역직각삼각형
N = 5
for i in range(N):
    for j in range(N - i):
        print("*", end="")
    print("")
```

```python
# 별찍기: 직각삼각형
N = 5
for i in range(N, -1, -1):
    print("*" * (i + 1))
```

```python
# 별찍기: 정삼각형
N = 5
for i in range(N):
    for j in range(N - i):
        print(" ", end="")
    for j in range(i * 2 + 1):
        print("*", end="")
    print("")
```

```python
# 별찍기: 정삼각형
N = 5
for i in range(N):
    print(" " * (N - i), "*" * (i * 2 + 1))
```

### List comprehension

```python
num = [1, 2, 3, 4]
result1 = []
for n in num:
    result1.append(n * 3)
print(result1)

# 위의 코드를 간단하게...
result2 = [n * 3 for n in num]
print(result2)

# 조건문 추가
result3 = [n * 3 for n in num if n % 2 == 0]
print(result3)
```
