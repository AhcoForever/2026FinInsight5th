### 딕셔너리

```python
# 다음 데이터는 무엇을 의미할까요?
psh = ["박성훈", 20, 'B']
```

```python
# 이름 = 박성훈
# 나이 = 20
# 혈액형 = B

# 데이터를 이렇게 명확하게 표현하면 더 좋겠죠?
```

```python
# 이렇게 속성(key) : 값(value) 대응으로 구성하는 자료형이
# 딕셔너리입니다

# {
#    "key": "value"
# }

psh = {
    "name": "박성훈",
    "age": 20,
    "blood": "B"
}
```

```python
# 숫자도 키로 사용 가능
d1 = {
    0: 1,
    1: 2
}

d2 = {
    'a': 11,
    'b': 22
}

# 각 요소에 접근은 [key]로 가능
print(d1[0])
print(d2['a'])
```

```python
# dict에 값 추가&삭제하기
d = {
    'name': 'psh',
    'age': 20
}

# 새로 추가할 키에 값 추가
d['blood'] = 'B'
print(d)

# 삭제하기
del d['age']
print(d)
```

```python
# 중복 키 값
d = {
    'a': 0,
    'a': [1, 2, 3]
}
print(d)
```

### 관련 함수

```python
# keys: dict 내 존재하는 key 리스트(dict_keys) 반환
d = {
    'name': 'psh',
    'age': 20,
    'blood': 'B'
}
print(d.keys())
print(list(d.keys()))
```

```python
# values: dict 내 존재하는 값 리스트(dict_values) 반환
d = {
    'name': 'psh',
    'age': 20,
    'blood': 'B'
}
print(d.values())
print(list(d.values()))
```

```python
# items: dict 내 존재하는 (key, value) 리스트(dict_items) 반환
d = {
    'name': 'psh',
    'age': 20,
    'blood': 'B'
}
print(d.items())
print(list(d.items()))
```

```python
# clear: dict 초기화
d = {
    'name': 'psh',
    'age': 20,
    'blood': 'B'
}
print(d)
d.clear()
print(d)
```

```python
# get: 해당 키 값의 value 가져오기
d = {
    'name': 'psh',
    'age': 20,
    'blood': 'B'
}
print(d.get('name'))
print(d['name'])
print(d.get('birth'))
print(d.get('birth', '0519'))
print(d['birth'])
```

```python
# 키 존재 여부
d = {
    'name': 'psh',
    'age': 20,
    'blood': 'B'
}

print('name' in d)
print('birth' in d)
```
