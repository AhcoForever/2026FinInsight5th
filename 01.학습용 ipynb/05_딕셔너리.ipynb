### 딕셔너리

```python
# 숫자도 키로 사용 가능
d1 = {
    0: 1,
    1: 2
}

d2 = {
    'a': 11,
    'b': 22
}

# 각 요소에 접근은 [key]로 가능
print(d1[0])
print(d2['a'])
```

```python
# 딕셔너리 중복 키 값
d = {
    'a' :0,
    'a' : [1,2,3]
}

print(d)
```

### 관련 함수

```python
# keys : dict 내 존재하는 key 리스트(dict_keys) 반환
d = {
    'name' : 'kmy',
    'age' : 20,
    'blood': 'B'
}
print(d.keys()) # 3.6 부터 dict key는 이렇게 나타나게되었다. 이전에는 아래 모양으로 떴었음.
print(list(d.keys()))
```

```python
# values: dict 내 존재하는 값 리스트(dict_values) 반환
print(d.values())
print(list(d.values()))

# 데이터 안에 특정 속성을 갖고 있는지 확인할 때 사용하기도함.
```

```python
# items : dict내 존재하는 (key, vlaues) 리스트(dict_items)반환
d = {
    'name': 'kmy',
    'age': 20,
    'blood': 'B'
}
print(d.items())
print(list(d.items()))
```

```python
# get: 해당 키 값의 value 가져오기
d = {
    'name' : 'kmy',
    'age' : 20,
    'blood': 'B'
}

print(d.get('name'))
print(d['name'])
print(d.get('birth'))
print(d.get('birth', '0404'))
print(d['birth'])

# 출력값 None : 없다!
```

```python
# 키 존재 여부
print('name' in d)
print('birth' in d)
```
