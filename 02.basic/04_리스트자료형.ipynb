### 리스트

```python
# 순서가 있는 데이터의 모음
# [1, 2, 3] 와 같이 [] 내 데이터 나열
# 첫번째부터 0 인덱스로 접근 가능
```

```python
l1 = []                 # 비어있는 리스트
l2 = [1, 2, 3]          # 3개의 값이 들어있는 리스트
l3 = [1, 'B', 3, 'D']   # 여러 데이터 타입이 들어가 있을 수도
l4 = [1, 2, [3, 4]]     # 리스트 내 리스트도 가능
```

```python
# 리스트의 길이
l = [1, 2, 3]
len(l)
```

```python
# 리스트의 인덱싱
l = [1, 2, 3]
l[0] + l[2] # 1 + 3
```

```python
# 리스트의 인덱싱
l = [1, 2, [3, 4]]
print(l[-1])
print(l[2])
print(l[2][0])
```

```python
# 리스트 슬라이싱
l = [1, 2, 3, 4, 5]
l[0:2] # 부분도 리스트
```

```python
# 리스트 슬라이싱
l = [1, 2, 3, 4, 5]
print(l[2:])
print(l[:2])
```

### 리스트 연산

```python
# 리스트 더하기
l1 = [1, 2, 3]
l2 = [4, 5, 6]
l1 + l2
```

```python
# 리스트 반복하기
l = [1, 2, 3]
l * 3
```

```python
# 리스트 반복하기: 0이 10개
l = [0] * 10
print(l)
```

```python
# 리스트 수정
l = [1, 2, 3]
l[2] = 4
print(l)
```

```python
# 리스트 요소 삭제
l = [1, 2, 3]
del l[2]
print(l)
```

```python
# 리스트 요소 삭제: 여러개
l = [1, 2, 3]
del l[1:]
print(l)
```

### 리스트 함수

```python
# append: 뒤에 값 추가하기
l = [1, 2, 3]
l.append(4)
print(l)
```

```python
# sort: 정렬하기: 오름차순
l = [5, 6, 3, 1]
l.sort()
print(l)
```

```python
# sort: 정렬하기: 내림차순
l = [5, 6, 3, 1]
l.sort(reverse = True)
print(l)
```

```python
# reverse: 순서 뒤집기
l = [1, 2, 3]
l.reverse()
print(l)
```

```python
# insert: 특정 위치에 값 삽입하기
l = [1, 3, 4]
# insert(i, V)
# i: index (2번째 >> 1)
# V: Value
l.insert(1, 2)
print(l)
```

```python
# index: 해당 값 존재시 index 값 반환
# 없으면 에러
l = [1, 2, 3]
print(l.index(1))
print(l.index(4))
```

```python
# remove: 해당 값 삭제(앞에서 부터 1개만)
l = [1, 2, 3, 1, 2, 3]
l.remove(1)
print(l)
```

```python
# pop: 맨 마지막 값 꺼내기
l = [1, 2, 3]
print(l.pop()) # 맨 마지막 요소 꺼내서 "보여준다"
print(l)
```

```python
# count: 요소 개수 세기
l = [1, 2, 3, 1, 2, 3]
l.count(1)
```

```python
# extend: 리스트 확장 (List + List 동일)
l = [1, 2, 3]
l.extend([4, 5])
print(l)
```

```python
# 리스트의 복사
l1 = [1, 2, 3]
l2 = l1

print("변경 전")
print(l1)
print(l2)
l1[0] = 4
print("변경 후")
print(l1)
print(l2)
print("=" * 10)
print(f"l1: {id(l1)}")
print(f"l2: {id(l2)}")
```

```python
# 리스트의 복사
l1 = [1, 2, 3]
l2 = l1[:]

print("변경 전")
print(l1)
print(l2)
l1[0] = 4
print("변경 후")
print(l1)
print(l2)
print("=" * 10)
print(f"l1: {id(l1)}")
print(f"l2: {id(l2)}")
```

```python
# 리스트의 복사
l1 = [1, 2, [3, 4]]
l2 = l1[:]

print("변경 전")
print(l1)
print(l2)
l1[2][0] = 4
print("변경 후")
print(l1)
print(l2)
print("=" * 10)
print(f"l1: {id(l1[2])}")
print(f"l2: {id(l2[2])}")
```
