### 연습문제1

```python
# 같은 글자가 몇개?

s = "starbucks"
TARGET = 's' 
count=0
for i in s:
    if i == TARGET:
        count+=1
print(count)
```

```python
s = "starbucks"
TARGET = 'bucks'
count = 0

for i in range(len(s)- len(TARGET)+1):
   if TARGET == s[i:i+len(TARGET)]:
      count+=1
print(f"result: {count}")
```

```python
# 모든 약수를 출력해라! 10이면 1,2,5,10 
TARGET = 100
result=[]
for i in range(1, TARGET+1):
    if (TARGET % i) == 0 :
        result.append(i)
print(result)
```

```python
# 약수를 찾는 다른 방법
TARGET = 100
result=[1]
for i in range(2, TARGET//2+1):
    if (TARGET % i) == 0 :
        result.append(i)
result.append(TARGET)
print(result)
```
