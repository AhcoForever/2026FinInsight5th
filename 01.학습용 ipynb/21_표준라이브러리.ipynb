### 표준 라이브러리
이미 만들어져 있는 라이브러리, 가져와서 활용하기

```python
# datetime: 날짜 및 시간 관련 라이브러리
import datetime

# data(년, 월, 일)
day1 = datetime.date(2025, 12, 25)
day2 = datetime.date(2026,1,1)

# 0: 월요일 ~ 6 : 일요일
print(day1.weekday())

# 1: 월요일 ~ 7: 일요일
print(day1.isoweekday())

# 날짜 차이
diff = day2 - day1
print(diff.days)

# 현 시간
print(datetime.datetime.now())

# datetime(년, 월, 일, 시, 분, 초)
KST = datetime.timezone(datetime.timedelta(hours=9))
dt1 = datetime.datetime(2025, 12, 25, tzinfo=KST)
dt2 = datetime.datetime(2025, 12, 25, tzinfo=datetime.timezone.utc)
print(dt1)
print(dt2)
```

```python
# time: 시간 관련 라이브러리
import time
# 1970년 1월1일 0시 0초 기준, 지난 초 반환
print(time.time())

# 연, 월, 일, 시, 분, 초 포맷으로 변환
print(time.localtime(time.time()))

# 특정 포맷으로 출력하기
print(time.strftime("%y-%m-%d", time.localtime(time.time())))

# 일시 중시 time.sleep(sec)
for _ in range(5):
    print("*")
    time.sleep(1)
```

```python
# random: 임의의 난수 생성
import random

# 0.0~1.0 사이 난수 [0.0 ,1.0]
print(random.random())

# 1~10 사이 정수 [1,10]
print(random.randint(1,10)) # 뒤에 숫자를 포함하지 않음 (10 불포함)

# 1~10 사이 정부 [1,10]
print(random.randrange(1,10)) # 뒤에 숫자를 포함함 (10 포함)

# 무작위로 섞기
a = [1,2,3,4,5,6,7,8,9]
random.shuffle(a)
print(a)

# 임의의 하나 선택하기
print(random.choice(a))

# 임의의 여러 개 선택하기
print(random.sample(a,3))
```

```python
# 로또 번호 랜덤 생성하기
# 첫번째 방법
import random
def get_lotto():
    result = []
    # 6개의 서로 다른 숫자 생성
    while(len(result)<=6):
        num = random.randint(1,46)
        if num not in result:
            result.append(num)
    result.sort
    return result
```

```python
# 로또 번호 랜덤 생성하기
# 두번째 방법
import random
def get_lotto():
    number = range(1,46)
    result = random.sample(number,6)
    result.sort()
    return result
print(get_lotto())
```

```python
# threading: 동시에 작업을 진행할 수 있도록 thread 활용
import time
import threading
def long_task(num): # 어떤 수행하는 일인데, 동시에 시키고 싶을때...
    for i in range(5):
        print(f"[{num}]: {i}")
        time.sleep(num+1)
print("start")
threads = []
for i in range(5):
    t = threading.Thread(target=long_task, args=(i,))
    threads.append(t)

for t in threads:
    t.start()
print("end")
```
