```python
num1 = 10
num2 = 1
if num1> num2 :
    print(f"num1: {num1}")
    print(f"num2 : {num2}")
    print(f"num1이 num2보다 {num1 - num2}만큼 큽니다.")

else:
    print("num1이 num2보다 크지 않네요")
```
