narry는 행렬로 계산해줌
- 모든 요소는 동일한 타입을 담는 다차원 배열
- 한번 정해지면 크기 변경 불가
- ndarray 클래스를 사용하여 생성

```python
# 1차원 배열 
import numpy as np

a1 = np.array([1,2,3])
print(a1)
print(a1.ndim) # 차원
print(a1.shape) # 모양
print(a1.size) # 크기
print(a1.dtype) # 데이터 타입
print("=" * 20)
```

```python
# 2차원 배열
a2 = np.array([[1.0, 2, 3], [4, 5, 6]])
print(a2)
print(a2.ndim)
print(a2.shape)
print(a2.size)
print(a2.dtype)

# 데이터 타입 우선순위 : 문자열 > 실수 > 정수 > 불린
# 배열에 하나라도 존재하면 데이터 타입이 우선순위에 따라 결정됨. 가장 포괄적인 타입을 선택함.
```

```python
# 3차원 배열
a3 = np.array([[[1,2],[3,4],[5,6],[7,8]]])
print(a3)
print(a3.ndim)
print(a3.shape)
print(a3.size)
print(a3.dtype)

# 단어 유사도를 구분할때 쓰임
# 사용자가 남긴 댓글을 3차원 배열에 넣음. 한 단어의 바로 옆 주소 있는건 서로 비슷한 단어로 구분을 할 수 있음
```

```python
# 배열 연산
a = np.array([1,2,3])
b = np.array([4,5,6])

print("배열의 합 ", a+b)
print("배열의 차 ", a-b)
print("배열의 곱", a*b)
print("스칼라 합", a+1)
```

```python
# 배열 연산

a = np.array([1,2,3])

print("합", a.sum())
print("평균", a.mean()) # 나누기가 있어서 출력값이 실수로 나온다.
print("최대값", a.max())
print("최소값", a.min())
```

```python
a = np.array([1,2,3,4,5])

# 인덱싱
b = a[0] # 1
b = a[2] # 3

# 슬라이싱
d = a[1:4] # [2,3,4]
e = a[:3] # [1,2,3]
```

```python
# 2차원 

a = np.array([[1,2,3],[4,5,6]])

# 인덱싱
b = a[0,0] # 1
c = a[1,2] # 6

# 슬라이싱
d = a[0,1:3] # [2,3]
e = a[:1] # 모든 줄에서 인덱스 1번째 요소를 모두 출력 [2,5]
f = a[:,:2] 
```

### 주요 함수

```python
# 배열 크기 정의 및 초기화
# 0으로 채워진 배열

zeros_arr = np.zeros((3,4))
print(zeros_arr)
print(zeros_arr.ndim)
print(zeros_arr.shape)
print(zeros_arr.size)
print(zeros_arr.dtype)

print("="*20)
# 1로 채워진 배열

ones_arr = np.ones((2,3), dtype = np.int16)
print(ones_arr)
print(ones_arr.ndim)
print(ones_arr.shape)
print(ones_arr.size)
print(ones_arr.dtype)

print("="*20)
# 임의의 값
empty_arr = np.empty((2,3)) # 배열안에 임의의 값이 들어간다. 비어있는 배열이 아니다!
print(empty_arr)
print(empty_arr.ndim)
print(empty_arr.shape)
print(empty_arr.size)
print(empty_ar.dtype)
```

```python
# arrange :  범위 내 간격 기준 나열
# 시작, 끝, 간격

a1 = np.arange(10, 30, 5)
print(a1)

a2 = np.arang(0,20.3)
print(a2)
```

```python
# linspace:  범위 내 수만큼 균등하게 나눔
# 시작, 끝, 나눌 수 
a1 = np.linespace(0,2,9)
print(a1)

a2 = np.linspace(1,2,10
print(a2))
```

```python
# 0과 1 사이의 균등 분포에서 난수를 생성하여 배열
arr = np.random.random((3,3))
print(arr)
```
